```python
import ee
import geemap

# Initialize
ee.Authenticate()
ee.Initialize(project='ee-ahobbs')

# Load your FeatureCollection
table = ee.FeatureCollection("projects/ee-ahobbs/assets/lsms_points")

# Create a geemap Map
m = geemap.Map()

# Add a clean basemap
m.add_basemap("CARTO_POSITRON")

# Style the points (must use only supported arguments)
point_style = {
    "color": "FF0000",     # outline color (hex string)
    "fillColor": "FF0000", # fill color
    "pointSize": 1         # pixel size
}

styled_points = table.style(**point_style)

# Add to map
m.addLayer(styled_points, {}, "LSMS Points")

# Center on your data
m.center_object(table, zoom=6)

m
```

```python
# Load your points
table = ee.FeatureCollection("projects/ee-ahobbs/assets/lsms_points")

# Create map
m = geemap.Map()

# Add a clean basemap
m.add_basemap("CARTO_POSITRON")

# --------------------------------------------------
# GLOBAL NDVI LAYER (LANDSAT COMPOSITES)
# --------------------------------------------------

dataset = (
    ee.ImageCollection("LANDSAT/COMPOSITES/C02/T1_L2_32DAY_NDVI")
    .filterDate("2017-01-01", "2017-12-31")
)

# Do NOT clip â†’ this allows global coverage
colorized = dataset.select("NDVI")

# NDVI visualization parameters
colorizedVis = {
    "min": 0,
    "max": 1,
    "palette": [
        'ffffff', 'ce7e45', 'df923d', 'f1b555', 'fcd163', '99b718', '74a901',
        '66a000', '529400', '3e8601', '207401', '056201', '004c00', '023b01',
        '012e01', '011d01', '011301'
    ]
}

# Add global NDVI
m.addLayer(colorized, colorizedVis, "Global NDVI Composite (2017)")


# --------------------------------------------------
# Points overlay
# --------------------------------------------------

point_style = {
    "color": "FF0000",
    "fillColor": "FF0000",
    "pointSize": 1
}

styled_points = table.style(**point_style)
m.addLayer(styled_points, {}, "LSMS Points")

# Center map on your points (NDVI stays global)
m.center_object(table, 6)

m
```
